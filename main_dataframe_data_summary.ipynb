```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define font sizes
SMALL_SIZE = 12
MEDIUM_SIZE = 14
BIGGER_SIZE = 16

# Set font sizes for different plot elements
plt.rc('font', size=SMALL_SIZE)          # default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # tick labels on x-axis
plt.rc('ytick', labelsize=SMALL_SIZE)    # tick labels on y-axis
plt.rc('legend', fontsize=SMALL_SIZE)    # legend
plt.rc('figure', titlesize=BIGGER_SIZE)  # figure title

# Set ggplot style
plt.style.use('ggplot')

# Load your dataset
data = pd.read_csv('./GAN/final_synthetic_gas_compositional_dataset.csv')

# Define the selected features
selected_features = ['T (K)', 'P (MPa)', 'methane', 'ethane',
                     'propane', 'butane', 'pentane', 'ipentane', 'hexane',
                     'heptane', 'octane', 'nonane', 'decane', 'helium', 'oxygen',
                     'nitrogen', 'water', 'argon','hydrogen',
                     'H2S', 'CO2', 'z']

# Create subplots for the features
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(15, 15))
fig.subplots_adjust(hspace=0.5)

# Loop through the selected features and create histograms
for i, feature in enumerate(selected_features):
    row, col = divmod(i, 3)
    ax = axes[row, col]
    ax.hist(data[feature], bins=20)  # Adjust bins as needed
    ax.set_title(f'Distribution of {feature}')
    ax.set_xlabel('')
    ax.set_ylabel('')

# Customize layout for the last subplot (if needed)
if i < 14:
    for j in range(i + 1, 15):
        row, col = divmod(j, 3)
        fig.delaxes(axes[row, col])

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create boxplots for the features
fig, axes = plt.subplots(nrows=8, ncols=3, figsize=(12, 12))
fig.subplots_adjust(hspace=0.5)

# Loop through the selected features and create boxplots
for i, feature in enumerate(selected_features):
    row, col = divmod(i, 3)
    ax = axes[row, col]
    data.boxplot(column=feature, ax=ax)
    ax.set_title(f'Boxplot of {feature}')
    ax.set_xlabel('')
    ax.set_ylabel('')

# Customize layout for the last subplot (if needed)
if i < 14:
    for j in range(i + 1, 15):
        row, col = divmod(j, 3)
        fig.delaxes(axes[row, col])

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Create a correlation heatmap for feature correlations
corr_matrix = data[selected_features].corr()
plt.figure(figsize=(16, 13))
sns.heatmap(corr_matrix, annot=True, cmap="RdYlGn")
plt.title("Correlation Heatmap of Selected Features")
plt.show()
```

```python
# Basic descriptive statistics of the dataset
summary_stats = data[selected_features].describe()

# Display the summary statistics
print("Summary Statistics of Selected Features:")
print(summary_stats)

```
