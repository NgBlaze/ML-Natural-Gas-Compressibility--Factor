```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'Pressure (PSI)': [5, 10, 15, 20, 25],
    'Wax Deposits': [284.21, 165.81, 152.99, 143.89, 125.44]
    
}

# Create DataFrame
df = pd.DataFrame(data)

# Scatter Plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Pressure (PSI)'], df['Wax Deposits'], color='blue', marker='o')
plt.title('Pressure vs Wax Deposits')
plt.xlabel('Pressure (PSI)')
plt.ylabel('Wax Deposits')
plt.grid(False)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
parameters = ['Kinematic Viscosity', 'Dynamic Viscosity']
values = [0.94, 0.782]  # Kinematic in cSt, Dynamic in cP

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(parameters, values, color=['blue', 'green'])

# Adding labels
plt.title('Viscosity Calculation')
plt.ylabel('Viscosity (cSt/cP)')

# Adding value labels
for i in range(len(values)):
    plt.text(i, values[i], f'{values[i]}', ha = 'center')

# Show plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data
data = {
    'P1 (Inlet) Sample B': [5,6,7, 8, 15, 20,23, 25],
    'P1 (Inlet) Sample A': [5,6,7, 8, 15, 20,23, 25],
    'P2 (Outlet) Sample B': [3.5,4.0,4.5, 8.8, 12.6, 19.0,22, 24.4],
    'P2 (Outlet) Sample A': [3.2,4.0,4.5, 8.6, 12.3, 18.6,22, 24.0],
    
    'Wax Deposits': [278.43,260,254.21, 160.52, 152.99, 146.77, 134.64, 116.32]
}

# Create DataFrame
df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))

# Plotting both pressures against wax deposits
plt.plot(df['Wax Deposits'], df['P1 (Inlet) Sample A'], label='Sample A', marker='o')
plt.plot(df['Wax Deposits'], df['P1 (Inlet) Sample B'], label='Sample B', marker='o')
plt.plot(df['Wax Deposits'], df['P2 (Outlet) Sample B'], label='Sample B', marker='o')
plt.plot(df['Wax Deposits'], df['P2 (Outlet) Sample B'], label='Sample B', marker='o')

# Adding labels and title
plt.title('Pressure of Both Crude Sample vs Wax Deposits')
plt.xlabel('Wax Deposits')
plt.ylabel('Pressure (PSI)')
plt.legend()
plt.grid(False)

# Show plot
plt.show()
```
