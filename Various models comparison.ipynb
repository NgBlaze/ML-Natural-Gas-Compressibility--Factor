```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the DataFrame from a CSV file
df = pd.read_csv("Various model_predictions.csv")

# Set a color palette for the regression lines
color_palette = sns.color_palette("Set1", n_colors=len(df.columns) - 1)

# Create a scatter plot with a regression line
plt.figure(figsize=(10, 6))

# Loop through the columns in the DataFrame (excluding 'z_Actual')
for i, column in enumerate(df.columns[1:], 1):
    sns.regplot(x='z_Actual', y=column, data=df,
                label=column, color=color_palette[i - 1],
                scatter_kws={'s': 100}, line_kws={'color': 'gray', 'linestyle': '--', 'linewidth': 1})

# Add labels and legend
plt.xlabel('Actual Values', fontsize=14)
plt.ylabel('Predictions', fontsize=14)
plt.title('Regression Plot of Model Predictions vs. Actual', fontsize=16)
plt.legend(fontsize=12)

# Customize the grid
plt.grid(True, linestyle='--', alpha=0.7)

# Set a white background
plt.gca().set_facecolor('white')

# Show the plot
plt.tight_layout()
plt.show()
```
